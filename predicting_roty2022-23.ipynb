```python
import pickle
import pandas as pd
import numpy as np
```

```python
file_name1 = 'FlaskAPI/models/model_file.p'
file_name2 = 'FlaskAPI/models/transform_file.p'
    
with open(file_name1, 'rb') as pickled:
    data = pickle.load(pickled)
    model = data['model']
        
with open(file_name2, 'rb') as pickled:
    data = pickle.load(pickled)
    transform = data['transform'] 
```

```python
df = pd.read_csv('cleaned_data2.csv', index_col=0)
df
```

```python
df_tbd = df[df['ROOKIE'] == 1]
df_tbd_quant = df_tbd.select_dtypes(include=np.number).drop("ROOKIE", axis=1)
```

```python
X = df_tbd_quant.drop("ROTY", axis=1)
```

```python
X = transform.transform(X)
```

```python
y_hat = model.predict_proba(X)
```

```python
df_tbd['ROTY'] = (1 - y_hat)
```

```python
df_tbd[['PLAYER', 'ROTY']].sort_values(by="ROTY", ascending=False)
```
