```python
import pandas as pd
```

```python
df_train = pd.read_csv('draft_data.csv', index_col=0)
```

```python
df_train.dropna(inplace=True)
i = df_train.loc[df_train.TEAM == "1.2"].index #dropped jason preston because he missed his rookie season and so his college stats loaded in
df_train = df_train.drop(i)                         # improperly
```

```python
df_train['CENTER'] = df_train['POSITION'].apply(lambda x: 1 if 'center' in x.lower() else 0)
df_train['FORWARD'] = df_train['POSITION'].apply(lambda x: 1 if 'forward' in x.lower() else 0)
df_train['GUARD'] = df_train['POSITION'].apply(lambda x: 1 if 'guard' in x.lower() else 0)
df_train['ROOKIE'] = 0
df_train['YEAR'] = df_train['YEAR'].apply(lambda x: x.split("-")[0])
```

```python
df_train
```

```python
df_pred = pd.read_csv('roty_pred.csv')
```

```python
df_pred.dropna(inplace=True)
```

```python
df_pred['CENTER'] = df_pred['POSITION'].apply(lambda x: 1 if 'center' in x.lower() else 0)
df_pred['FORWARD'] = df_pred['POSITION'].apply(lambda x: 1 if 'forward' in x.lower() else 0)
df_pred['GUARD'] = df_pred['POSITION'].apply(lambda x: 1 if 'guard' in x.lower() else 0)
df_pred['ROOKIE'] = 1
df_pred['ROTY'] = 'NaN'
```

```python
df_pred
```

```python
draft_df_cleaned = pd.concat((df_train, df_pred), ignore_index=True, join='inner')
```

```python
draft_df_cleaned.rename(columns = {'FG%':'FGPercent', '3PM':'ThreePM', '3PA':'ThreePA', '3P%':'ThreePPercent', \
                                   'FT%':'FTPercent', '+/-':'PlusMinus'}, inplace=True)
```

```python
draft_df_cleaned
```

```python
draft_df_cleaned.to_csv("draft_df_cleaned.csv")
```
